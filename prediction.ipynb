```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import pickle
import pandas as pd
import numpy as np
```

```python
#load models
model = load_model('models/model.h5')

with open('pickles/onehotencoder.pkl', 'rb') as f:
    onehotencoder = pickle.load(f)

with open('pickles/labelencoder.pkl', 'rb') as f:
    labelencoder = pickle.load(f)

with open('pickles/scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)
```

```python
input_data = {
    'CreditScore': 516,
    'Geography': 'Germany',
    'Gender': 'Female',
    'Age': 21,
    'Tenure' : 1,
    'Balance' : 600,
    'NumOfProducts': 1,
    'HasCrCard': 0,
    'IsActiveMember': 1,
    'EstimatedSalary': 500000,
}
```

```python
#pre-processing of data encode categorical data, and Geography col - OneHotEncoding
geo_encoder = onehotencoder.transform([[input_data['Geography']]])
geo_encoder_df = pd.DataFrame(geo_encoder, columns=onehotencoder.get_feature_names_out(['Geography']))
geo_encoder_df
```

```python
input_df = pd.DataFrame([input_data])
input_df
```

```python
input_df['Gender'] = labelencoder.fit_transform(input_df['Gender'])
input_df
```

```python
#concate the categorical and geography cols into final_df
input_df = pd.concat([input_df.drop('Geography', axis=1), geo_encoder_df], axis=1)
```

```python
#scalling the dataset
scaled_df = scaler.transform(input_df)
scaled_df
```

```python
#prediction

prediction = model.predict(scaled_df)

prediction
```

```python
prediction_prob = prediction[0][0]
```

```python
if prediction_prob > 0.5:
  print("Customer is likely to leave the bank")
else:
  print("Customer is likely to stay with the bank")
```
